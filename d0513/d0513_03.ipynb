```python
import pandas as pd
df = pd.read_csv('../z20_data/drinks.csv')
df
```

```python
# NA로 기재된 정규데이터 처리
df.fillna('NN',inplace=True)
```

```python
df
```

### 2. agg : 그룹함수(['함수명',....나열])

```python
## 대륙별 (1개 컬럼 컬럼선택 하여) 최소, 최대, 평균, 합계
df.groupby('continent')['beer_servings'].min()
df.groupby('continent')['beer_servings'].max()
df.groupby('continent')['beer_servings'].sum()
df.groupby('continent')['beer_servings'].mean()
```

```python
df.groupby('continent')['beer_servings'].agg(['min','max','mean','sum'])
df.groupby('continent')['spirit_servings'].agg(['min','max','mean','sum'])
```

```python
df.groupby('continent')[['beer_servings','spirit_servings']].sum()
```

```python
# 각 컬럼당 agg가 각각 적용
df.groupby('continent')[['beer_servings','spirit_servings','wine_servings']].agg(['min','max','mean','sum'])
```

### 3. 조건문 사용

```python
df.groupby('continent').count()
```

```python
df.groupby('continent').value_counts()
```

```python
# 대륙별로 맥주소비량 합계
df.groupby('continent')['beer_servings'].sum()
```

```python
# 전체 평균보다 맥주 소비량이 높은 대륙을 출력
# 1) 전체 평균 소비량  : 106만병
df['beer_servings'].mean()
# 2) 대륙별 평균 소비량
df.groupby('continent')['beer_servings'].mean()

# 변수화
beer_mean = df['beer_servings'].mean()
continent_over_mean = df.groupby('continent')['beer_servings'].mean()

# 전체 평균보다 맥주 소비량이 높은 대륙 검색  -> 어디서 비교를 하여 출력해야 하는지 정해서 테이블 지정
continent_over_mean[continent_over_mean >= beer_mean]
```

```python
# 간단한 형태의 로직 참고
# df.groupby('학년')['국어'].mean()
# df['국어'].mean()
# kor_groupby = df.groupby('학년')['국어'].mean()
# kor_groupby[kor_groupby > df['국어'].mean()]

# 전체 알코올 합계 평균
df['total_literes_of_pure_alcohol'].mean()
total_groupby = df.groupby('continent')['totla_litres_of>pure_alcohol'].mean()
total_groupby[total_groupby > df['totla_litres_of>pure_alcohol'].mean]
```

```python
# total_litres_of_pure_alcohol 전체 평균보다 높은 대륙을 출력
df['total_litres_of_pure_alcohol'].mean()  # 4.72
df.groupby('continent')['total_litres_of_pure_alcohol'].mean()

tlpa_total = df['total_litres_of_pure_alcohol'].mean() 
tlpa_continent = df.groupby('continent')['total_litres_of_pure_alcohol'].mean()
tlpa_continent[tlpa_continent>=tlpa_total]
```

```python
# 대륙 AF에 해당되는 국가, 맥주소비량만 출력 : 아래의 2가지가 같다
df[df['continent']=='AF'][['country','beer_servings']]
df.groupby('continent').value_counts().loc['AF']

# 퍼센트를 구하기 위해서는 groupby를 사용
df.groupby('continent').value_counts(normalize=True).loc['AF']
# 그 중에서 특정 컬럼만 선택
df.groupby('continent')['beer_servings'].value_counts(normalize=True).loc['AF']
len(df.groupby('continent')['beer_servings'].value_counts(normalize=True).loc['AF'])
df_groupby = df.groupby('continent')['beer_servings'].value_counts(normalize=True).loc['AF']
df_groupby['beer-servings']
```

```python
# 대륙별 카운트 중 AF의 퍼센트 출력
df['continent'].count()
df.groupby('continent')['beer_servings'].value_counts(normalize=True).loc['AF']
(df.groupby('continent')['beer_servings'].value_counts(normalize=True).loc['AF']*100).astype(str)+"%"
```
