```python
import pandas as pd
data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
df
```

```python
## 5) 평점에 숫자 "점" 이라는 글자를 추가해서 출력 ex) 9.2점

### a> 타입변환
# df['평점'] = df['평점'].astype(str)  # 타입변환
df['평점'] = df['평점'].astype(str)+"점"
# df['평점'] = "평점 " + df['평점'].astype(str)
df
```

```python
### b> 함수적용
def txt_ad(rating):  # 매개변수 rating = int
    return  str(rating) + "점"  # 값이 되도록
    
```

```python
df['평점'] = df['평점'].apply(txt_ad)
df
```

```python
### c> 람다함수
df['평점'] = df['평점'].apply(lambda rating:str(rating)+"점")
df
```

```python
## 4) 개봉 연도 역순정렬 출력
df['개봉 연도'].sort_values(ascending=True)
df.sort_values('개봉 연도',ascending=True)
```

```python
## 3) 추천 점수 컬럼을 생성해서 (관객수 * 평점) // 100
# 추천점수 컬럼 역순정렬
# df['추천 점수'] =  ((df['관객 수'] * df['평점'])/100).astype(float)
df['추천 점수'] =  (df['관객 수'] * df['평점'])//100
df
```

```python
df['추천 점수'].sort_values(ascending=True)
df
```

```python
## 2) 2015년 이후 개봉한 영화 데이터에서 영화, 개봉연도만 출력
filt = df['개봉 연도'] > 2015
df[filt]
df.loc[filt,['영화','개봉 연도']]
# df[filt,['영화','개봉 연도']]  -> loc를 빼면 error
```
