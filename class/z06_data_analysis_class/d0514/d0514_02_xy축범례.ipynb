### x,y축의 설명문구

```python
import matplotlib.pyplot as plt
import matplotlib

# 한글설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # : windows 사용자용
# matplotlib.rcParams['font.family'] = 'HYGungSo-Bold'  # : windows 사용자용
matplotlib.rcParams['font.size'] = '15' 
matplotlib.rcParams['axes.unicode_minus'] = False 

x = ['han','kim','hong']
y = [20,40,80]
y2 = [50,30,90]
y3 = [80,90,95]

# 범례 이름지정  : label
plt.plot(x,y,label='국어')
plt.plot(x,y2,label='영어')
plt.plot(x,y3,label='수학')
# 범례 표시, 위치 설정
plt.legend(loc=1,ncols=3)
# plt.legend(loc='upper left')

# 그래프 이름
plt.title('막대그래프',fontdict = {'family':'Malgun Gothic','size':30})

# x,y축의 설명문구, 위치, 색상
plt.xlabel('이름',loc='right',color='red') # left/center/right
plt.ylabel('국어',loc='top',color='green') # top/center/bottom
```
