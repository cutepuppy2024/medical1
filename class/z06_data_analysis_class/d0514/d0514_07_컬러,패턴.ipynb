```python
import matplotlib.pyplot as plt
import matplotlib

# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # : windows 사용자용
matplotlib.rcParams['font.size'] = '15' 
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df.head(3)
```

### 막대별 컬러지정

```python
# 국, 영, 수, 사, 과
x = df.head(5)['이름']
y1= df.head(5)['국어']
y2 = df.head(5)['영어']
y3 = df.head(5)['수학']
y4 = df.head(5)['과학']
y5 = df.head(5)['사회']
```

```python
colors = ['r','g','b','y','pink']
plt.plot(x,y1,label='국어')
plt.plot(x,y2,label='영어',marker='o',linestyle='--')
plt.plot(x,y3,label='수학',marker='s', ls='-.')
plt.plot(x,y4,label='과학',marker='x', ls=':')
plt.plot(x,y5,label='사회',marker='D')

for i, txt in enumerate(y3):
    plt.text(x[i],y3[i]+0.9,txt,ha='center')

plt.legend(loc='upper center', ncols=5)
plt.show()
```

```python
colors = ['r','g','b','y','pink']
plt.bar(x,y1,label='국어',color= colors)
# plt.plot(x,y2,label='영어',marker='o',linestyle='--')
# plt.plot(x,y3,label='수학',marker='s', ls='-.')
# plt.plot(x,y4,label='과학',marker='x', ls=':')
# plt.plot(x,y5,label='사회',marker='D')

for i, txt in enumerate(y3):
    plt.text(x[i],y3[i]+0.9,txt,ha='center')

plt.legend(loc='upper right')
plt.show()

```
