```python
import matplotlib.pyplot as plt
import matplotlib

# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # : windows 사용자용
matplotlib.rcParams['font.size'] = '15' 
```

```python
x = [1,2,3]
y = [12,9,8]

# 그래프 그리기, 선두께 : linewidth=5 label : 범례
# marker='o'/'x'
# markersize = ms : 마커 크기
# markeredgecolor = mec : 마커테두리색상
# markerfacecolor = mfc : 마커색상
# alpha : 투명도

# plt.plot(x,y,linewidth=5, marker='o', markersize= 10, markeredgecolor='red',markerfacecolor='blue', label='국어')  
plt.plot(x,y,linewidth=5, marker='o', ms= 10, mec='red', mfc='yellow', label='국어', alpha=0.7, ls=':')  # 약어사용 가능
# plt.legend(loc='upper right')
plt.legend(loc=(0.2,0.5)) # 위치지점 변경
plt.title('막대그래프')
# y축 표시 간격을 조정
plt.yticks([5,7,9,11,13,15])
plt.xticks([1,2,3])
```
