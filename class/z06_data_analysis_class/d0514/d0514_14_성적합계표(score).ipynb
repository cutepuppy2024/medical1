```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd


# 한글설정
matplotlib.rcParams['axes.unicode_minus'] = False 
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # : windows 사용자용
matplotlib.rcParams['font.size'] = '15' 
```

```python
# 막대그래프 score.xlsx 합계점수를 뽑아서 그래프로 표현하시오
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
df['합계'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df
```

```python
x = df['이름']
y = df['합계'].values

plt.bar(x,y,label='합계')

plt.grid(axis='y',alpha=0.3,linestyle='--',linewidth=2, color='purple')
plt.legend(loc='upper right')
plt.xticks(rotation=90)
plt.title('성적합계표')

plt.xlabel('이름',loc='right',color='purple')
plt.ylabel('합계',loc='top',color='Red')
plt.show()
```
