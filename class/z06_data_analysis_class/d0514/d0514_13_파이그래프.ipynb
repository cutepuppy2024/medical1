```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd

# 한글설정
matplotlib.rcParams['axes.unicode_minus'] = False 
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # : windows 사용자용
matplotlib.rcParams['font.size'] = '15' 
```

```python
values = [30,25,20,13,10,2]

labels = ['Python','Java','Javascript','C#','C','ETC']
explode = [0.2]*6
colors = ['#edf041','#81db3b','#93db3b','#709bdb','#214bdb','#100dbb']
wedgeprops={'width':0.6, 'edgecolor':'w', 'linewidth':2}
```

```python
# plt.pie(values,labels=labels, autopct='%1.f%%',startangle=90,colors=colors, counterclock=False
#         ,wedgeprops=wedgeprops)
plt.pie(values,labels=labels, autopct='%1.f%%', startangle=90, colors=colors, explode = explode, counterclock=False
        ,wedgeprops=wedgeprops,pctdistance=0.7)
plt.legend(loc=(1.0,0.3))
plt.show()
```
