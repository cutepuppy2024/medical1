```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd


# 한글설정
matplotlib.rcParams['axes.unicode_minus'] = False 
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # : windows 사용자용
matplotlib.rcParams['font.size'] = '15' 
```

```python
values = [30,25,20,13,10,2]
# values = [1,1,1,1,1,2]
labels = ['Python','Java','Javascript','C#','C','ETC']
explode = [0.2,0,0,0,0,0]
# explode = [0.1]*6

# 시작위치 : startangle  /  회전방향 : counterlock= True/False
# plt.pie(values,labels=title, autopct='%.1f%%',startangle=90)
plt.pie(values,labels=labels, autopct='%.1f%%',startangle=90,explode =explode, counterclock=False)
plt.title('전체 언어별 선호도')
plt.legend(loc=(1.2,0.3),title='언어별 선호도')
plt.show()
```
