```python
import matplotlib.pyplot as plt
import matplotlib

# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # : windows 사용자용
matplotlib.rcParams['font.size'] = '15' 

x = [1,2,3]
y = [2,4,8]
```

```python
# --, -., :, default는 실선

plt.plot(x,y,label='국어',linestyle=':',marker='o')  
plt.legend(loc='upper center')
plt.show()
```

```python
plt.plot(x,y,label='국어',linestyle=':',marker='o')  
plt.legend(loc='upper center')

for i, txt in enumerate(y):
    plt.text(x[i],y[i]+0.5,txt,ha='center',color='blue')
    
# plt.text()  : 좌표에 데이터 2,4,8 의 값을 출력
# ha='center' : 가운데 정렬
# y[i]+0.5 : 위치값 조정
    
plt.yticks([0,1,2,3,4,5,6,7,8])
# plt.yticks([0:10])
    
# 보여지는 그래프 전체크기가 커짐, 화면 해상도 조절
plt.figure(dpi=200)

# 저장 : show 위에
plt.savefig('g.png')
plt.savefig('g2.png',dpi=300) # 화질이 좋아지면서 커진다, 용량도 커짐

plt.show()
```

```python
# plt.plot(x,y,label='국어')
plt.bar(x,y,label='국어')
plt.legend(loc='upper center')

for i, txt in enumerate(y):
    plt.text(x[i],y[i]+0.5,txt,ha='center',color='blue')
    
# plt.yticks([0,1,2,3,4,5,6,7])
# plt.yticks([0:10])
plt.yticks([x for x in range(0,11)])

# 보여지는 그래프 전체크기가 커짐, 화면 해상도 조절
plt.figure(dpi=200)

# 저장 : show 위에
plt.savefig('g.png')
plt.savefig('g2.png',dpi=300) # 화질이 좋아지면서 커진다, 용량도 커짐

plt.show()
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
# 키 그래프 출력

x = df['이름']
y = df['키']

# plt.plot(x,y)
# plt.bar(x,y)

plt.bar(x,y,label='키') 
plt.legend(loc='upper right')

for i,txt in enumerate(y):
    plt.text(x[i],y[i]+1.5,txt,ha='center', color='blue')

# y축 표시 : 170,180,190,200,210
plt.yticks([165,170,180,190,200,210])
plt.xticks(rotation=90)
plt.ylim(165,210)
plt.title('키 그래프')
```

```python
# 키 그래프 옆으로 막대 그래프  :  barh 

x = df['이름']
y = df['키']

# plt.plot(x,y)
# plt.bar(x,y)

plt.barh(x,y,label='키')  # 약어사용 가능
plt.legend(loc='upper right')

for i,txt in enumerate(y):
    plt.text(y[i]+1.5,x[i],txt,ha='center', color='blue')

plt.xticks([165,170,180,190,200,210])
plt.xlim(165,210)
plt.title('키 그래프')
```
