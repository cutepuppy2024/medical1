```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
def grade(score):
    if score >= 90 : return 'A'
    elif score >= 80 : return 'B'
    elif score>=70 : return 'C'
    elif score >=60 : return 'D'
    else : return 'F'
```

```python
# 평균 컬럼을 추가해서 grade컬럼에 A >=90,B>=80,C>=70,D>=60,F<60 학점을 출력
df['합계'] = df['국어'] + df['영어'] +df['수학'] +df['과학'] +df['사회']
df
```

```python
df['평균'] = df['합계']/5
df
```

```python
df['grade'] = df['평균'].apply(grade)
df
```
