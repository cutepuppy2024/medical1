```python
def grade(score) :
    if score >= 90 :return "A"
    elif score >= 80 :return "B"
    elif score >= 70 :return "C"
    elif score >= 60 :return "D"
    else: return "F"
```

```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df['평균'] = df['총합'] /5
df
```

```python
df['평점'] = df['평균'].apply(grade)
df
```

### lambda 함수 : 익명함수 -> lambda 매개변수

```python
## 결과 컬럼 평균 70점 이상인 학생은 합격, 불합격
# df['결과'] = df['평균'].apply(함수)
df['결과'] = df['평균'].apply(lambda score:'합격' if score>= 70 else "불합격")
df
```

```python
# 65점 초과 합격, 65점 재시험, a미만은 불합격

df['결과'] = df['평균'].apply(lambda score:'합격' if score>= 65 else ("재시험" if score==65 else "불합격"))
df
```

```python
# notnull : null이 아닌지 확인  /   isnull  : null인지 확인

def lower_change(lang):
    if pd.notnull(lang) : return lang.lower()
    else : return "python"
```
