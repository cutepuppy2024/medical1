## 데이터 수정 정리

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
# columns 추가 : 없는 컬럼에 데이터를 입력하며 추가가 됨

df['합계'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df['결과'] = '불합격'
df
```

```python
# columns 수정

df['학교'] = df['학교'] + '등학교'
df
```

```python
df['국어'] = df['국어'] -10
df
```

```python
# 학교 이름 수정
df['학교'].replace({"구로고등학교":"디지털고","단지고등학교":"영등포고"},inplace=True)
df
```

```python
df['SW특기'] = df['SW특기'].str.upper()
df
```

```python
# 컬럼부분 삭제
df.drop(columns='사회',inplace=True)
df
```

```python
df.drop(columns=['국어','과학'],inplace=True)
df
```

```python
# values 출력시 리스트 내의 리스트로 출력

df.values
```

```python
# for문 사용시 배열(리스트)로 출력됨

for df in df.values:
    print(df)
```
