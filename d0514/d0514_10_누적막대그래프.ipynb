```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd


# 한글설정
# matplotlib.rcParams['axes.unicode_minus'] = False 
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # : windows 사용자용
matplotlib.rcParams['font.size'] = '15' 
```

```python
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

### 누적막대그래프  : 더해져서 그래프가 그려짐

```python
x = df['이름']
y1 = df['국어']
y2 = df['영어']
y3 = df['수학']
```

```python

plt.bar(x,y1,label='국어')
plt.bar(x,y2,bottom=y1,label='영어')
plt.bar(x,y3,bottom=y1+y2,label='수학')


for i,txt in enumerate(y1):
    plt.text(x[i],y1[i]-12,txt,ha='center')
    
for i, xt in enumerate(y2):
    plt.text(x[i],y1[i]+y2[i]-15,txt,ha='center')
    
for i,txt in enumerate(y3):
    plt.text(x[i],y1[i]+y2[i]+y3[i]-11,txt,ha='center')
    
plt.legend(loc='upper center')

plt.show()
```
