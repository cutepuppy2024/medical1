```python
# 합을 구하는 방법
df['0~4세'] + df['5~9세'] + df['10~14세']
df.iloc[0,0] + df.iloc[0,1] + df.iloc[0,2]
df.iloc[0,:3].sum()
df.iloc[0,3:13].sum()
df.iloc[0,13:].sum()
```

```python
x = ['2023년1월']
y1 = df.iloc[0,:3].sum()
y2 = df.iloc[0,3:13].sum()
y3 = df.iloc[0,13:].sum()
```

```python
plt.bar(x,y1,label='0~14세')
plt.bar(x,y2,bottom=y1,label='15~64세')
plt.bar(x,y3,bottom=y1+y2,label='65세 이상')

for i,txt in enumerate(y1):
    plt.text(x[i],y1[i],txt,ha='center')
for i,txt in enumerate(y2):
    plt.text(x[i],y2[i],txt,ha='center')
for i,txt in enumerate(y3):
    plt.text(x[i],y3[i],txt,ha='center')
    
plt.legend(loc='upper center')

plt.show()
```
