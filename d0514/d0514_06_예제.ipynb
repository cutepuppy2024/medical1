```python
import matplotlib.pyplot as plt
import matplotlib

# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # : windows 사용자용
matplotlib.rcParams['font.size'] = '15' 
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
# 합계 컬럼을 생성해서 막대그래프, 옆으로 누운 그래프를 출력
# 합계 점수를 출력

df['합계'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] +df['사회']
df
```

```python
x = df['이름']
y = df['합계']

plt.barh(x,y,label='키')

for i,txt in enumerate(y):
    plt.text(y[i]+1.5,x[i],txt,ha='center',color='blue')
plt.xlim(0,500)
plt.title('합계그래프')
plt.legend(loc='upper center')
```
