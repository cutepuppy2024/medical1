```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # : windows 사용자용
matplotlib.rcParams['font.size'] = '15' 
```

```python
import pandas as pd
df = pd.read_csv('../z20_data/stu_score1.csv')
h_df = df.head(20)
```

```python
x = df['id'].head(20)
y = df['total'].head(20)
y1 = df['kor'].head(20)
y2 = df['eng'].head(20)
y3 = df['math'].head(20)
```

```python
plt.bar(x,y)
plt.ylim(150,300)
plt.xticks(rotation=90)

plt.show()
```

```python
idx = np.arange(h_df.shape[0])
idx
```

```python
plt.figure(figsize=(15,5))
# plt.bar(np_arr,y1)
plt.bar(idx-0.3,y1,width=0.3)
plt.bar(idx,y2,width=0.3)
for i,txt in enumerate(y2):
    plt.text(idx[i],y2[i]+0.5,txt,ha='center')
plt.bar(idx+0.3,y3,width=0.3)
plt.xticks(idx,x,rotation=90)

plt.show()
```
