```python
# chipotle.tsv에서 주문 중 가장 많이 주문한 메뉴 10개 출력하여 꺾은선 그래프로 출력
```

```python
import pandas as pd
df = pd.read_csv('../z20_data/chipotle.tsv',sep='\t')
df
```

```python
# item_name 별 주문한 횟수가 몇 번인지  
df['item_name'].value_counts()[:10]
```

```python
y = df['item_name'].value_counts()[:10].values
x = df['item_name'].value_counts()[:10].index

plt.bar(x,y)
plt.plot(x,y)
```

```python
# item_name 별 주문한 개수의 합  
# 주문별 그룹
# quantity의 합을 구해야 함
df.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(5)
```

```python
x = df.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(5).index
y = df.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(5).values

plt.bar(x,y)
plt.plot(x,y)
```

```python
x = df['item_name'].value_counts()[:5].index
y = df['item_name'].value_counts()[:5].values
y2 = df.groupby('item_name')['quantity'].sum()[:5].sort_values(ascending=False).values
plt.plot(x,y)
plt.plot(x,y2)
```

```python
plt.figure(dpi=200) # 화면 늘리기
plt.xticks(rotation=90)
plt.plot(x,y,label='주문수')
plt.plot(x,y2,label='주문횟수')
plt.legend(loc='upper right')
```
