```python
# diabetes.csv LinearRegression, RandomForest
```

```python
df = pd.read_csv('diabetes.csv')
df
```

```python
df.info()
df.shape
df.describe()
```

```python
data = df.iloc[:,1:]
target = df.iloc[:,-1]
```

```python
train_input, test_input, train_target, test_target = train_test_split(
    data, target,random_state=42
)
```

```python
# 표준화
from sklearn.preprocessing import StandardScale
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input, train_target)
test_scaled = ss.transform(test_input, test_target)
```

```python
# LinearRegression
lr = LinearRegression()
lr.fit(train_scaled)
```

```python
print(lr.score(train_scaled, train_scaled))
print(lr.score(test_scaled, test_target))
```

```python
# RandomForest
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_jobs = 1,random_state=42)
scores = cross_validate(rf,train_scaled, train_target, return_train_score=True, n_jobs=-1)
print(np.mean(scores['train_score']),np.mean(scores['test_score']))
```
