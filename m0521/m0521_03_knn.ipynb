```python
from sklearn import svm, metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import glob, os.path, re
import random

matplotlib.rcParams['axes.unicode_minus'] = False 
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # : windows 사용자용
matplotlib.rcParams['font.size'] = '10' 
```

```python
# np 구조로 변환
perch_length = np.array([8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0, 
     21.0, 21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5, 
     22.5, 22.7, 23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5, 
     27.3, 27.5, 27.5, 27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0, 
     36.5, 36.0, 37.0, 37.0, 39.0, 39.0, 39.0, 40.0, 40.0, 40.0, 
     40.0, 42.0, 43.0, 43.0, 43.5, 44.0])

perch_weight = np.array([5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 
     110.0, 115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 
     130.0, 150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 
     197.0, 218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 
     514.0, 556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 
     820.0, 850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 
     1000.0, 1000.0])
```

```python
# 그래프 구현
plt.scatter(perch_length, perch_weight)

# 0의 데이터가 있는지 확인, 직선의 모양이 어떤지/ 그래프가 직선or곡선에 가까운지 확인
```

```python
# 1. 데이터 전처리 - train,test 세트 구성
# random_state = 42  :  랜덤으로 섞는 것을 고정 -> 예측률이 고정됨   /   실제에서는 쓰지 않는다 -> 계속 결과값을 산출해보고 좋은 값의 데이터를 저장하여 사용
train_input, test_input, train_target, test_target = train_test_split(perch_length, perch_weight, random_state=42)
```

```python
# train_input - 2가지 이상의 행렬로 구성, 2차원 행렬로 구성
# 1차원 train_input, test_input -> 2차원의 행렬로 변경
# 2개 이상의 특성값을 가지면 2차원 핼렬이 되어 reshape 할 필요가 없음.
train_input = train_input.reshape(-1,1)
train_input.shape
test_input = test_input.reshape(-1,1)
```

```python
# 2. 데이터 훈련시키기
knr = KNeighborsRegressor()  # knn회귀 - 예측 알고리즘을 사용
knr.fit(train_input, train_target)
```

```python
# 3. 예측값 구하기
predict = knr.predict(test_input)
print("예측값 :",predict)
```

```python
plt.scatter(test_input, test_target)    # 실제값
plt.scatter(test_input, predict)        # 예측값 /  plot은 값이 일률적이지 않게 그려짐
```

```python
# train < test   : 과소적합   ->    n_neighbors의 값을 줄이기
# train 예측률
score = knr.score(test_input, test_target)
print("test 예측률 :", score)
# test 예측률 
score2 = knr.score(train_input, train_target)
print("train 예측률 :", score2)
```

```python
knr.n_neighbors = 3   # 5 -> 3
knr.fit(train_input, train_target)
```

```python
# n_neighbors 개수를 5 -> 3개로 줄임
# train 예측률
score = knr.score(test_input, test_target)
print("test 예측률 :", score)
# test 예측률 
score2 = knr.score(train_input, train_target)
print("train 예측률 :", score2)
```
