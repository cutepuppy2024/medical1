### one_hot_encoding

```python
df.head()
```

```python
# 종류는 컬럼별로 카운트
# 실제 컬럼은 22개 -> 문자의 다른 컬럼들은 12개
```

```python
[0]*12
```

```python
# v = 0일 때, 22개의 데이터가 넘어옴
# x	s	n	t	p	f	c	n	k	...	s	w	w	p	w	o	p	k	s	u
```

```python
target = [] 
data = []
attr_list = []
for row_index, row in df.iterrows() :
    target.append(row.loc[0])    # target 결과값
    row_data = []
    for col, v in enumerate(row.loc[1:]):
        if row_index == 0:
            attr = {"dic":{},"cnt":0}  # cnt : 컬럼별 다른 문자의 수 -> 12개
            attr_list.append(attr)
        else:
            attr = attr_list[col]   
        
        # 원핫인코딩 변환    
        d = [0]*12
        if v in attr["dic"]:
            idx = attr["dic"][v]
            # print(idx)
        else:
            idx =  attr['cnt']
            attr['dic'][v] = idx
            # print(attr['dic'][v])
            attr["cnt"] += 1
             
        d[idx] = 1
        row_data += d            # 원핫인코딩의 데이터 저장파일 -데이터의 가중치 때문에 정답률이 다를 수 있음
        
    data.append(row_data)        # 1줄 리스트를 다시 리스트에 추가
data[0]
    
# df2 = pd.DataFrame(data)
# df2
```
