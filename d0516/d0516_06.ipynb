### 2. 형변환 replac 사용 - 남,녀 데이터  : df, df2

```python
# 남자데이터
df = pd.read_excel('../z20_data/20230104_연령별인구현황.xlsx',skiprows=3, usecols='B,E:Y',index_col = '행정기관')
df.head(1)
```

```python
df.index
```

```python
# df.loc['전국  ']
df.iloc[0]
```

```python
df.columns
```
