```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
import numpy as np
df.loc['3번','영어'] = np.nan
df.loc['4번','수학'] = np.nan
df.loc['5번',['국어','영어','수학']] = np.nan
df
```

```python
## 총합 컬럼 만들어서, 국어, 영어, 수학, 과학, 사회 합계 점수를 총합에 입력

df['총합'] = 0
df
```

```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회'] 
df
```

```python
df['총합'].sum()
```

```python
# 데이터전처리 : NaN -> 0
df.fillna(0,inplace=True)
```

```python
df.info()
```

```python
# 타입변경 후 출력이 아니라 입력
df['국어'] = df['국어'].astype(int)
```

```python
df['국어'].mean()
```

```python
df['결과'] = '불합격'
df
```

```python
## 총합 300 이상이면 합격, 아니면 불합격 / 400 이상이면 불합격 
filt = df['총합'] > 300
# df[filt]
df[df['총합'] > 300]
# df.loc[df['총합'] > 300,'결과'] = '합격'
df.loc[filt,'결과'] = '합격'

df.loc[df['총합'] > 400,'결과'] = '우수'
df


# df
```

```python
## 총합과 결과를 사회 뒤쪽으로 옮겨서 출력

a_list= list(df.columns)
a_list

df= df[a_list[0:7] + a_list[9:-1] + [a_list[8]]]
df
```
