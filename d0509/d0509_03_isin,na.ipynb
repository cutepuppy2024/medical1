```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
# 2개이상 조건 검색부분
lang = ['Python','Java']
filt = df['SW특기'].isin(lang)
df[filt]
```

```python
# 문자열 검색 함수  : startswith, endswith, contains, lower, upper
# is  함수 : isin
```

```python
# 문자열함수 적용 - lower(), upper()
lang = ['python','java']
df['SW특기'].str.lower()
df['SW특기'].str.upper()

# isin 명령어를 사용해서 2개 이상의 검색을 할 수 있도록 함. 
filt = df['SW특기'].str.lower().isin(lang)
df[filt]
```

```python
# java라는 글자가 포함되어 있는 것을 검색
# error : NaN 값 때문 -> 확인!!

filt = df['SW특기'].str.contains('Java')

df[filt]  
```

```python
# Na 처리  : na=True/False
filt = df['SW특기'].str.contains('Java',na=True)

filt
df[filt]
```

```python
filt = df['SW특기'].str.contains('Java',na=False)

filt
df[filt]
```

```python
filt = df['SW특기'].str.lower().str.contains('java',na=True)
df[filt]
```

```python
filt = df['SW특기'].str.lower().str.contains('java',na=False)
df[filt]
```

```python
import pandas as pd
df = pd.read_csv('../z20_data/stu_score1.csv',index_col='no')
df
```
