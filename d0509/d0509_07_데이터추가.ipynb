### 데이터 수정 

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
# 1) 선택한 컬럼의 값을 모두 변경

df['학교'] = '디지털고'
df
```

```python
# 2) 선택한 컬럼의 특정 값을 변경  : replace({"":""})

df['학교'].replace({'구로고':'디지털고'},inplace=True)
df
```

```python
## SW특기 컬럼에서 java가 포함되어 있는 열을 출력
find = df['SW특기'].str.lower().str.contains('java',na=False)
df[find]
```

```python
# 3) 소문자/대문자로 변경  : lower/upper()

df['SW특기'] = df['SW특기'].str.lower()
df['SW특기'] = df['SW특기'].str.upper()
df['SW특기'] 
```

```python
# 4) NaN의 값을 변경  : fillna()
df.fillna("미선택",inplace=True)
df
```

```python
# 5) 문자열 추가  : +
df['학교'] = df['학교'] + '등학교'
df
```

```python
## 과학점수를 +5점 더해서 출력
# df.info()
df['과학'] = df['과학'] + 5
df
```

```python
## 국어점수를 +10 점 더해서 출력
df['국어'] = df['국어'] +10
df
```
