```python
import pandas as pd

df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
## 사회 컬럼 삭제, 음악 컬럼 추가 
## 음악점수는 과학점수가 80점 이상이면 80점, 미만이면 65점으로 입력 
## 합계 컬럼 추가하여  국어+영어+수학+과학+음악
## 컬럼의 위치는 음악이 과학 바로 뒤로 변경
```

```python
## 사회 컬럼 삭제, 음악 컬럼 추가 
df.drop(columns='사회',inplace=True)
df['음악'] = 65
df
```

```python
## 음악점수는 과학점수가 80점 이상이면 80점, 미만이면 65점으로 입력 
filt = df['과학'] >=80 
df[filt]

df.loc[df[filt].index,'음악'] = 80
df
```

```python
## 컬럼의 위치는 음악이 과학 바로 뒤로 변경

a_list = list(df.columns)
a_list
df[a_list[0:7] + [a_list[-1]] + [a_list[7]]]
```

```python
df.info()
```

```python
## 합계 컬럼 추가하여  
df['합계'] = 0

# df.iloc[0,3]

sum = 0
# 국어+영어+수학+과학+음악
for i in range(len(df)):
    for j in range(3,8):
        sum += df.iloc[i,j]   # i가 index 0일때 -> j는 3~7  [0,3],[0,4],[0,5],[0,6],[0,7]을 더함

    df.iloc[i,-1] = sum       # 위치값  i가 0일때의 합계에 sum을 넣는다
df


# => 미완성 
```
