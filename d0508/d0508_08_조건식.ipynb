```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

### 데이터 선택 - 조건

```python
# pandas 조건문
filt = df['키'] > 190
df[filt]
```

```python
# True에 해당하는 값만 DataFrame(df)로 출력
df[df['키'] > 190]
```

```python
df.loc[['1번','6번']]
df.iloc[[0,7]]
```

```python
# 키가 평균보다 큰 학생을 출력하시오
filt = df['키'] > df['키'].mean()  # 조건에 맞는 row를 찾음
df.loc[filt,'키']                  # DataFrame 형태로 출력 
df.loc[filt,['이름','키']]   
df.loc[~filt,['이름','키']]        
```

```python
# 국어 85점 이상, 영어 85점 이상 이름, 학교, 국어, 영어 출력하시오

# and : & , or : | , not : ~ => 2가지 조건일 경우 ()를 꼭 !!

# 둘 중 하나라도 85점 이상인 학생
filt1 = df['국어'] >= 85 
filt2 = df['영어'] >= 85

df.loc[filt1 | filt2,['이름','학교','국어','영어']]

# 둘 다 85점 이상인 학생
filt = (df['국어'] >= 85 ) & (df['영어'] >= 85)
df.loc[filt,['이름','학교','국어','영어']]
```
